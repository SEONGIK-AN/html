### HTML 시작

- HTML은 HyperText Markup Language의 약자로 **HTML은 콘텐츠의 구조를 정의하는 마크업 언어**
- HTML에 사용되는 태그는 보통 닫는 태그를 함께 사용하지만, 단일으로 사용하는 태그도 존재함.
<pre>
  1. &lt;br&gt;
  2. &lt;img&gt;
  3. &lt;hr&gt;
</pre>

```python
html=\
'''
<!DOCTYPE html>
<html lang="ko">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
  </head>
  <body>
    Hello, World!
  </body>
</html>
'''
disp(html)
```

#### 서식

```python
html=\
'''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <!-- 다음은 강조 효과를 주는 태그이다. -->
    <p><em>약한 강조</em>는 &lt;em&gt; <strong>강한 강조는</strong> &lt;strong&gt;을 쓴다.</p>
    <p>&lt;b&gt;는 글씨를 <b>굵게</b> 해준다.</p>
    <p>&lt;mark&gt;는 글씨에 <mark>하이라이팅</mark> 표시를 해준다.</p>
    <hr />
    <!-- 다음은 밑줄 또는 취소선을 주는 태그이다. -->
    <p>&lt;ins&gt;는 문서의 <ins>삽입(insert)</ins>을 표시하는 태그로 밑줄을 표시해준다</p>
    <p>&lt;u&gt;는 <u>밑줄(underline)</u>을 표시하는 태그로 밑줄을 표시해준다. 의미 상으로는 철자 오류를 표기할 때 쓴다.</p>
    <p>&lt;del&gt;은 문서의 <del>삭제(delete)</del>를 표시하는 태그로 취소선을 표시해준다</p>
    <p>&lt;s&gt;은  <s>취소선(strikethough)</s>을 표시하는 태그로 약한 삭제를 의미한다.</p>
    <hr />
    <!-- 다음은 글자크기에 영향을 주는 태그이다. -->
    <p>&lt;sup&gt;는 <sup>윗첨자</sup>를 나타내준다.</p>
    <p>&lt;sub&gt;는 <sub>아랫첨자</sub>를 나타내준다.</p>
    <p>&lt;small&gt;은 글씨를 <small>작게</small> 표시해주며, 보통 저작권 정보나 주석을 표기할 때 쓴다.</p>
    <hr />
    <!-- 다음은 인용구 관련 태그이다. -->
    <p>&lt;q&gt;는 <q>인라인</q>으로 인용을 나타내준다.</p>
    <p>&lt;blockquote&gt;는 블록단위의 인용을 나타내준다.</p>
    <blockquote>이것은 인용된 문구이다.</blockquote>
    <p>&lt;abbr&gt;은 축약을 표시해준다.</p>
    <abbr title="HyperText Markup Language">HTML</abbr>
    <p>&lt;address&gt;은 주소를 표시해준다.</p>
    <address>서울특별시 강남구</address>
    
</body>
</html>
'''
disp(html)
```

### Attribute

#### Global Attribute

|    | Attribute       | 의미                                                                          |
|---:|:----------------|:------------------------------------------------------------------------------|
|  0 | class           | 요소에 대해 하나 이상의 클래스 이름을 지정(style 지정 등에 사용)              |
|  1 | contenteditable | 요소 내용을 편집할 수 있는지 여부를 지정                                      |
|  2 | data-*          | 페이지나 애플리케이션에 비공개로 사용자 정의 데이터를 저장하는 데 사용        |
|  3 | dir             | 텍스트 방향 지정                                                              |
|  4 | draggable       | 요소를 드래그할 수 있는지 여부를 지정, 링크와 이미지는 기본적으로 드래그 가능 |
|  5 | enterkeyhint    | 가상 키보드의 Enter 버튼에 어떤 글자를 넣을지 지정                            |
|  6 | hidden          | 요소를 숨길 때 사용                                                           |
|  7 | id              | 요소의 고유 ID를 지정, class와 달리 유일값이 보장됨                           |
|  8 | inert           | 요소를 비활성화함(표시는 되나, 요소 사용이 안됨)                              |
|  9 | inputmode       | 가상 키보드의 키보드 모드 변경 (Ex. numeric, email)                           |
| 10 | lang            | 컨텐츠의 언어 지정                                                            |
| 11 | popover         | 다른 컨텐츠 위에 표시되도록 지정                                              |
| 12 | spellcheck      | 요소의 철자 및 문법 검사 여부를 지정                                          |
| 13 | style           | 요소의 인라인 스타일을 지정                                                   |
| 14 | tabindex        | 요소의 탐색 순서 지정(탭으로 탐색)                                            |
| 15 | title           | 요소에 대한 추가 정보를 지정, 보통은 툴팁으로 표시되는 텍스트                 |

```python
html=\
'''
<!DOCTYPE html>
<html lang="ko">
  <head>
    <style>
      p.class_test1 {
        color: red;
      }
      p.class_test2 {
        color: blue;
      }
    </style>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Document</title>
  </head>
  <body>
    <h2>class</h2>
    <p class="class_test1">class를 지정하면</p>
    <p class="class_test2">이렇게</p>
    <p>스타일을 따로 적용할 수 있다.</p>
    <hr />
    <h2>contenteditable</h2>
    <p contenteditable="true">편집 가능한 요소</p>
    <p contenteditable="false">편집 불가능한 요소</p>
    <hr />
    <h2>dir</h2>
    <p dir="auto">이것은 auto</p>
    <p dir="ltr">이것은 ltr</p>
    <p dir="rtl">이것은 rtl</p>
    <hr />
    <h2>draggable</h2>
    <a href="" draggable="true">이건 옮길 수 있는 요소</a>
    <hr />
    <h2>hidden</h2>
    <p>제가 보이시나요?</p>
    <p hidden>제가 보이시나요?</p>
    <p hidden="true">제가 보이시나요?</p>
    <hr />
    <div inert>
      <button>누를 수 없는 버튼</button><br />
      <input>쓸 수 없는 블록</input><br />
    </div>
    <hr />
    <h2>popover</h2>
    <div>
      <h1 popover id="sayhello">Hello</h1>
      <button popovertarget="sayhello">Click</button>
    </div>
    <hr />
    <h2>spellcheck</h2>
    spellcheck: <input type="text" spellcheck="true" value="Is this sentence incorract? "></input>
    <hr />
    <h2>title</h2>
    <p title="title">마우스를 가져다 대보세요.</p>
  </body>
</html>
'''
disp(html)
```

### 공간분할

```python
html=\
'''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <li>&lt;p&gt;, &lt;div&gt;, &lt;h&gt;, &lt;ul&gt;, &lt;ol&gt;, &lt;form&gt;요소는 display 속성값이 블록(block)인 대표적인 요소</li>
    <h3 style="border: 1px solid gray;">이렇게</h3>
    <p style="border: 1px solid gray;">한 블록씩</p>
    <li style="border: 1px solid gray;">묶을 수 있다</li>
    <div style="border: 1px solid gray;">
        <h3>블록으로도</h3>
        <p>묶을 수 있다</p>
    </div>
    <hr />   
    <li>&lt;span&gt;으로 묶으면 특정 부분(inline)만 스타일을 적용할 수 있다</li>
    <p><span style="color: red">span으로 묶은 곳</span>은 이렇게 표시된다</p>
    <p><strong style="color: red">강조 요소</strong>에도 사용할 수 있다</p>
</body>
</html>
'''
disp(html)
```

### 레이아웃

- &lt;header&gt;: 보통 페이지 또는 섹션의 가장 윗부분
- &lt;nav&gt;: navigation의 약자로, 일반적으로 사이트를 안내하는 요소로 사용
- &lt;main&gt;: 문서의 주된 컨텐츠, 하나의 요소만 보이도록 해야함
- &lt;article&gt;: 웹 페이지의 내용에 사용하는 태그, 사이트 독립적으로 배포나 재사용 할 수 있는 섹션에 사용
- &lt;section&gt;: 웹 페이지 섹션에 사용하는 태그, 의미적으로 파트를 나눌 때 사용
- &lt;aside&gt;: 본문의 주요 부분을 표시하고 남은 부분 설명; 사이드바, 광고창 등
- &lt;footer&gt;: 페이지나 섹션의 가장 아랫부분; 라이선스, 주소, 연락처 등

#### DIV요소 사용

```python
html=\
'''
<!DOCTYPE html>
<html lang="ko">
<head>
    <style>
        div {
            border: none;
            padding-top: 1.2rem;
            padding-bottom: 1.2rem;
        }
        h2 {
            font-weight: bold;
            text-align: center;
            vertical-align: middle;
        }
        #header {
            background-color: #276ea1;
            width: 100%;
            height: 20%;
            color: white;
        }
        #nav {
            background-color: #26d8a2;
            width: 20%;
            height: 60%;
            float: left;
        }
        #main {
            background-color: #ffffff;
            width: 60%;
            height: 60%;
            float: left;
        }
        #aside {
            background-color: #ca6af7;
            width: 20%;
            height: 60%;
            float: left;
        }
        #footer {
            background-color: #eaec5b;
            width: 100%;
            height: 20%;
        }
    </style>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <div id="header"><h2>HEADER</h2></div>
    <div id="nav"><h2>NAV</h2></div>
    <div id="main"><h2>MAIN</h2></div>
    <div id="aside"><h2>ASIDE</h2></div>
    <div id="footer"><h2>FOOTER</h2></div>
</body>
</html>
'''
disp(html)
```

TABLE 요소 사용

```python
html=\
'''
<!DOCTYPE html>
<html lang="ko">
<head>
    <style>
        table {
            border-collapse: collapse;
            width: 100%;
            border: none;
            table-layout: fixed;
        }
        tr, td {
            border-collapse: collapse;
            border: none;
        }
        h2 {
            font-weight: bold;
            text-align: center;
            vertical-align: middle;
        }
        #header {
            background-color: #276ea1;
            color: white;
        }
        #nav {
            background-color: #26d8a2;
        }
        #main {
            background-color: #ffffff;
        }
        #aside {
            background-color: #ca6af7;
        }
        #footer {
            background-color: #eaec5b;
        }
    </style>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
</head>
<body>
    <table>
        <tr>
            <td id="header" colspan="5"><h2>HEADER</h2></td>
        </tr>
        <tr>
            <td id="nav"><h2>NAV</h2></td>
            <td id="main" colspan="3"><h2>main</h2></td>
            <td id="aside" ><h2>aside</h2></td>
        </tr>
        <tr>
            <td id="footer" colspan="5"><h2>FOOTER</h2></td>
        </tr>
    </table>
</body>
</html>
'''
disp(html)
```
